```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url_yk = "https://raw.githubusercontent.com/rahmadsa/dataset/main/transaksi_ritel_yk.csv"
url_mgl = "https://raw.githubusercontent.com/rahmadsa/dataset/main/transaksi_ritel_mgl.csv"

df_yk = pd.read_csv(url_yk)
df_mgl = pd.read_csv(url_mgl)

print(df_yk.head())
```

```python
df = pd.concat([df_yk, df_mgl], ignore_index=True)

print("Jumlah baris:", len(df))
print("Jumlah kolom:", len(df.columns))
```

```python
df['harga'].describe()
```

```python
Q1 = df['harga'].quantile(0.25)
Q3 = df['harga'].quantile(0.75)

IQR = Q3 - Q1

batas_bawah = Q1 - (1.5 * IQR)
batas_atas = Q3 + (1.5 * IQR)

print("Batas bawah :", batas_bawah)
print("Batas atas :", batas_atas)
```

```python
outlier = df[
    (df['harga'] < batas_bawah) |
    (df['harga'] > batas_atas)
]

print("Jumlah outlier:", len(outlier))
```

```python
df_no_outlier = df[
    (df['harga'] >= batas_bawah) &
    (df['harga'] <= batas_atas)
]

print("Jumlah data sebelum hapus outlier :", len(df))
print("Jumlah data sesudah hapus outlier :", len(df_no_outlier))
```

```python
df_no_outlier.isnull().sum()
```

```python
df_no_outlier.duplicated().sum()
```

```python
df_dedup = df_no_outlier.drop_duplicates()

print("Jumlah data sebelum deduplikasi :", len(df_no_outlier))
print("Jumlah data sesudah deduplikasi :", len(df_dedup))
```
